```python
import os
import json
import glob
from typing import Dict
import pandas as pd
```

```python
csv_cols = ["video_id", "category_id"]
df_cols = csv_cols + ["country", "category"]
data_dir = "src/"
```

```python
original_directory = os.getcwd()
try:
    os.chdir(data_dir)
except FileNotFoundError:
    os.chdir("..")
    original_directory = os.getcwd()
    os.chdir(data_dir)

df: pd.DataFrame = pd.DataFrame(columns=df_cols)

for csv_name in glob.glob("*.csv"):
    country: str = csv_name[:2]
    csv_data: pd.DataFrame = pd.read_csv(csv_name, usecols=csv_cols)
    csv_data["country"] = country

    with open(f"{country}_category_id.json", "r") as json_file:
        json_data = json.load(json_file)

    category_dict: Dict[int, str] = {}
    for video_category in json_data['items']:
        category_dict[int(video_category['id'])] = video_category['snippet']['title']
    if 29 not in category_dict:
        category_dict[29] = "Nonprofits & Activism"

    csv_data["category"] = csv_data["category_id"].apply(lambda x: category_dict[x])
#     print(len(csv_data))

    df = pd.concat([df, csv_data])

# del country, csv_data, csv_name, json_file, json_data, video_category, category_dict

os.chdir(original_directory)
del original_directory
```

```python
df.set_index(pd.RangeIndex(start=0, stop=df.shape[0], step=1), inplace=True)
```
