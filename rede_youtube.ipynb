```python
import os
import json
import glob
from typing import Dict
import pandas as pd
```

```python
csv_cols = ["video_id", "category_id", "views", "likes" , "dislikes", "comment_count",
            "trending_date", "comments_disabled", "ratings_disabled"]
ytdata_cols = csv_cols + ["country", "category"]
data_dir = "src/"
```

```python
original_directory = os.getcwd()
try:
    os.chdir(data_dir)
except FileNotFoundError:
    os.chdir("..")
    original_directory = os.getcwd()
    os.chdir(data_dir)

ytdata: pd.DataFrame = pd.DataFrame(columns=ytdata_cols)

for csv_name in glob.glob("*.csv"):
    country: str = csv_name[:2]
    csv_data: pd.DataFrame = pd.read_csv(csv_name, usecols=csv_cols)
    csv_data["country"] = country

    with open(f"{country}_category_id.json", "r") as json_file:
        json_data = json.load(json_file)

    category_dict: Dict[int, str] = {}
    for video_category in json_data['items']:
        category_dict[int(video_category['id'])] = video_category['snippet']['title']
    if 29 not in category_dict:
        category_dict[29] = "Nonprofits & Activism"
        
    csv_data["category"] = csv_data["category_id"].apply(lambda x: category_dict[x])

    ytdata = pd.concat([ytdata, csv_data])

del country, csv_data, csv_name, json_file, json_data, video_category, category_dict

os.chdir(original_directory)
del original_directory
```

```python
variaveis = pd.read_excel("vars_fake.xlsx")
variaveis = variaveis.set_index("var")
paises = variaveis.columns
gml = """graph [
    directed 0

"""
for i in range(variaveis.shape[1]):
    gml += f"""    node [
        id "{paises[i]}"
        label "{paises[i]}"

"""
    for j in range(variaveis.shape[0]):
        var = variaveis.index[j]
        gml += f"""        {var} {variaveis.loc[var][i]}
"""
    gml += "    ]\n"
conexoes_paises = pd.read_excel("edges_fake.xlsx")
conexoes_paises = conexoes_paises.set_index("paises")
weights = {}
for pais_linha in paises:
    for pais_coluna in paises:
        if (pais_linha, pais_coluna) not in weights and (pais_coluna, pais_linha) not in weights and pais_linha != pais_coluna:
            weights[pais_linha, pais_coluna] = conexoes_paises.loc[pais_linha][pais_coluna]
thresh = 10
for [source, target], weight in weights.items():
    if weight > thresh:
        gml += f"""    edge [
        source "{source}"
        target "{target}"
        weight {weight}
    ]
"""
gml += "]"
with open("rede.gml", "w") as file:
    file.write(gml)
```
